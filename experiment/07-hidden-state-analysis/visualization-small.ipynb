```python
import matplotlib.collections as clt
import ptitprince as pt
import scipy.stats as stats
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

type_list = ['active_promoter','strong_enhancer','txn_transition','txn_elongation','insulator','heterochrom_lo','repressed','repetitive_CNV']
output_path = 'results/'
```

```python
type = type_list[0]
type_array = np.load(output_path + type + '_small.npy')
not_type_array = np.load(output_path + type + '_not_small.npy')
print(type_array.shape)
print(not_type_array.shape)

df = pd.DataFrame(type_array, columns=['score'])
df['group'] = 'group1'
df_not = pd.DataFrame(not_type_array, columns=['score'])
df_not['group'] = 'group2'

data = pd.concat([df,df_not])

y_max=6
y_min=-7

pvalue = stats.mannwhitneyu(df['score'].dropna(),df_not['score'])[1]
print(pvalue)

dx="group"; dy="score"; ort="h"; pal="Reds_r"; sigma=.2
f, ax=plt.subplots(figsize=(3, 3))
plt.ylim((-7, 10))
plt.xlim((-1, 1))
pt.RainCloud(x=dx, y=dy, data=data, palette=pal, bw=sigma, box_sym='', rain_alpha = .1,
            width_viol = .4, ax = ax)

plt.plot([0,1], [y_max+(y_max-y_min)/10,y_max+(y_max-y_min)/10], linewidth=1, color='k') 
plt.plot([0,0], [y_max+(y_max-y_min)/10,y_max+(y_max-y_min)/10-(y_max-y_min)/20], linewidth=1, color='k') 
plt.plot([1,1], [y_max+(y_max-y_min)/10,y_max+(y_max-y_min)/10-(y_max-y_min)/20], linewidth=1, color='k') 

plt.text(0.2,y_max+(y_max-y_min)/8,f'P<0.0005',fontsize=10,color='k')

plt.xlabel("")
plt.ylabel("hidden state")
plt.xticks([0,1],['DNA in region','DNA not in region'])
plt.title('Active Promoter (Small Model)')
plt.savefig('images/raincloud-small-' + type + '.png',dpi=300, bbox_inches = 'tight')
plt.show()
```
