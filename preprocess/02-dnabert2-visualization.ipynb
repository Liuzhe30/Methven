```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
sns.set()
```

```python
embedding1 = np.load('dnabert1_test.npy')
embedding2 = np.load('dnabert2_test.npy')

d = np.random.normal(.4,2,(2,2))
norm = mcolors.TwoSlopeNorm(vmin=d.min(), vmax = d.max(), vcenter=0)

plt.figure(figsize=(25, 2))
ax = sns.heatmap(embedding1, vmin=-5, vmax=5,cmap=plt.cm.RdBu)
plt.ylabel('embedding1')
plt.show()

plt.figure(figsize=(25, 2))
ax = sns.heatmap(embedding2, vmin=-5, vmax=5,cmap=plt.cm.RdBu)
plt.ylabel('embedding2')
plt.show()
```

```python
plt.figure(figsize=(25, 2))
ax = sns.heatmap(embedding2-embedding1, vmin=-5, vmax=5,cmap=plt.cm.RdBu)
plt.ylabel('embedding2-embedding1')
plt.show()
```
