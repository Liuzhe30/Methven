```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
snp_list = ['rs3806624', 'rs7731626', 'rs2234067','rs2233424','rs947474','rs3824660','rs968567','rs3218251']
file_path = 'data/pred_merged/'
```

```python
# small model
data = pd.read_csv('data/pred_merged/merged_small.csv')

ax = plt.subplots(figsize=(3.2,6))

morandi_palette = ['#CD5C5C', '#D3D3D3','#000000', '#6495ED']
ax = sns.barplot(y = 'SNP', x = 'count of CpG sites', data = data, hue = 'type', ci=50,palette=morandi_palette,alpha=1)
plt.xlim(0, 50)

for patch in ax.patches:
    patch.set_edgecolor('black')
    patch.set_linewidth(1)

#plt.xlim(0, 36)
plt.legend(loc="best")
plt.title('effect of RA-associated SNPs\n(Small Model)')
plt.savefig('images/small.png',dpi=300, bbox_inches = 'tight')
plt.show()
```
