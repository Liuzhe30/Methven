```python
import pandas as pd
import numpy as np
import seaborn as sns
from prettytable import PrettyTable
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
balanced_path = '../../datasets/labeled_Mono/'

fig = plt.figure(figsize=(14,3))
data = pd.DataFrame()

for i in range(22):
    chr = 'chr' + str(i+1)
    table1 = pd.read_pickle(balanced_path + chr + '_small.dataset')
    table2 = pd.read_pickle(balanced_path + chr + '_large.dataset')

    num1 = table1.shape[0]
    num2 = table2.shape[0]

    data = data._append({'chr':chr, 'group':'1-10kbp','value count':num1},ignore_index=True)
    data = data._append({'chr':chr, 'group':'10k-100kbp','value count':num2},ignore_index=True)

#print(data.head())
color=['#F4A7C1','#3A4B6E']
sns.barplot(x='chr', y='value count', hue='group', data=data, palette=sns.set_palette(color)) 
plt.title('Distribution of samples (origin)', fontsize=15)
plt.savefig('images/sample-count1.png',dpi=300, bbox_inches = 'tight')
plt.show()
```
