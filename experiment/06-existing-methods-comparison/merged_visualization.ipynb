```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('results/merged_small.csv')

ax = plt.subplots(figsize=(7.5,3))
sns.barplot(x = 'metric', y = 'value', data = data, hue = 'model', ci=50,palette="gnuplot2",alpha=0.4)

plt.ylim(0.5, 1)
plt.title('Existing Methods Comparison\n (Small Model, 10-fold cross validation)')
plt.legend(loc="lower left")
plt.savefig('images/small.png',dpi=300, bbox_inches = 'tight')
plt.show()
```
