```python
import pandas as pd
import numpy as np
import seaborn as sns
from prettytable import PrettyTable
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
balanced_path = '../datasets/balanced/'

fig = plt.figure(figsize=(14,3))
data = pd.DataFrame()

for i in range(22):
    chr = 'chr' + str(i+1)
    table1 = pd.read_pickle(balanced_path + chr + '_small.dataset')
    table2 = pd.read_pickle(balanced_path + chr + '_middle.dataset')
    table3 = pd.read_pickle(balanced_path + chr + '_large.dataset')
    table4 = pd.read_pickle(balanced_path + chr + '_huge.dataset')

    num1 = table1.shape[0]
    num2 = table2.shape[0]
    num3 = table3.shape[0]
    num4 = table4.shape[0]

    data = data._append({'chr':chr, 'group':'1-1kbp','value count':num1},ignore_index=True)
    data = data._append({'chr':chr, 'group':'1k-10kbp','value count':num2},ignore_index=True)
    data = data._append({'chr':chr, 'group':'10k-100kbp','value count':num3},ignore_index=True)
    data = data._append({'chr':chr, 'group':'100k-1Mbp','value count':num4},ignore_index=True)

#print(data.head())
sns.barplot(x='chr', y='value count', hue='group', data=data,palette='bwr') 
plt.title('Distribution of samples before chromosome downsampling', fontsize=15)
plt.savefig('images/sample-count.png',dpi=300, bbox_inches = 'tight')
plt.show()
```

```python
balanced_path = '../datasets/chr_downsampled/'

fig, ax = plt.subplots(figsize=(14,3))
data = pd.DataFrame()

for i in range(22):
    chr = 'chr' + str(i+1)
    table1 = pd.read_pickle(balanced_path + chr + '_small.dataset')
    table2 = pd.read_pickle(balanced_path + chr + '_middle.dataset')
    table3 = pd.read_pickle(balanced_path + chr + '_large.dataset')
    table4 = pd.read_pickle(balanced_path + chr + '_huge.dataset')

    num1 = table1.shape[0]
    num2 = table2.shape[0]
    num3 = table3.shape[0]
    num4 = table4.shape[0]

    data = data._append({'chr':chr, 'group':'1-1kbp','value count':num1},ignore_index=True)
    data = data._append({'chr':chr, 'group':'1k-10kbp','value count':num2},ignore_index=True)
    data = data._append({'chr':chr, 'group':'10k-100kbp','value count':num3},ignore_index=True)
    data = data._append({'chr':chr, 'group':'100k-1Mbp','value count':num4},ignore_index=True)

#print(data.head())
sns.barplot(x='chr', y='value count', hue='group', data=data,palette='bwr') 
plt.title('Distribution of samples after chromosome downsampling', fontsize=15)
ax.get_legend().remove()
plt.savefig('images/sample-count-after.png',dpi=300, bbox_inches = 'tight')
plt.show()
```
