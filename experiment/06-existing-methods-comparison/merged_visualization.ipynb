```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

morandi_palette = ['#8A352E', '#304E7E', '#37939A']
```

```python
data = pd.read_csv('results/merged_small.csv')

ax = plt.subplots(figsize=(7.5,3))
sns.barplot(x = 'metric', y = 'value', data = data, hue = 'model', capsize = .05, ci=50,palette=morandi_palette,alpha=0.8)

plt.ylim(0.5, 1)
plt.title('Existing Methods Comparison\n (Small Model, 10-fold cross validation)')
plt.legend(loc="lower left")
plt.savefig('images/small.png',dpi=300, bbox_inches = 'tight')
plt.show()
```

```python
data = pd.read_csv('results/merged_small.csv')

markers = ["o", "*", "s"]
linestyles = ["-", "--", "-."]

ax = plt.subplots(figsize=(7.5,3))
sns.pointplot(
    x='metric', y='value', data=data, hue='model', palette=morandi_palette, 
    alpha=0.6, markers=markers, linestyles=linestyles, errwidth=2, capsize=0.05, ci=0,size=10, 
)

plt.ylim(0.65, 0.95)
plt.title('Existing Methods Comparison\n (Small Model, 10-fold cross validation)')
plt.legend(loc="lower left")
plt.savefig('images/small-2.png',dpi=300, bbox_inches = 'tight')
plt.show()
```
